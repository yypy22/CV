```python
plt.imshow(img)
plt.show()
```

```python
im_rgb = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(im_rgb)
plt.show()
```

```python
#Q2 make it Grayscale(0.2126R+0.7152G+0.0722B)

im_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(im_gray)
plt.show()
```

```python
#Binarization

gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
thresh, img_none = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
plt.figure(figsize=(12, 3))
plt.subplot(1,3,2)
plt.imshow(gray, cmap='gray')
plt.show()
plt.figure(figsize=(12, 3))
plt.subplot(1,3,3)
plt.imshow(img_none, cmap='gray')
plt.show()
```

```python
plt.hist(gray.ravel(), bins=250, rwidth=1.0, range=(0,250))
plt.text(0,250,'class0', fontsize=15)
plt.text(220,250,'class1', fontsize=15)
plt.xlabel('p i x e l')
plt.vlines(125, 0, 600, color='green', linestyles='dotted')
plt.show()
```
